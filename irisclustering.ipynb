```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


# Load the dataset
data = pd.read_csv("IRIS.csv")
data.head()
```

```python
clustering = KMeans(n_clusters=3)
clustering.fit(data.drop("species", axis=1))
data["Cluster"] = clustering.fit_predict(data.drop("species", axis=1))
centroids = clustering.cluster_centers_
print("Centroids:")
print(centroids)
```

```python
plt.scatter(data["sepal_length"], data["sepal_width"], c=data["Cluster"], cmap='rainbow', alpha=0.5, s=100)
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=100, marker='X', label='Centroids')
plt.title("KMeans Clustering of Iris Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
plt.show()
```

```python
plt.scatter(data["petal_length"], data["petal_width"], c=data["Cluster"], cmap='rainbow', alpha=0.5, s=100)
plt.scatter(centroids[:, 2], centroids[:, 3], c='black', s=100, marker='X', label='Centroids')
plt.title("KMeans Clustering of Iris Dataset")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.legend()
plt.show()
```
