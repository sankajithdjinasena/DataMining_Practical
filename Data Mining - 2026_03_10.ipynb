```python
li = ["Department of Software Engineeering in faculty of Computing",
      "Department of Data Science in faculty of Computing",
      "Department of Computing and Information Systems in faculty of Computing"]

dic = {}
for i in li:
  for key in i.split():
    if key in dic:
      dic[key] += 1
    else:
      dic[key] = 1

print(dic)
```

```python
feature_names = tfidf.get_feature_names_out()
print(feature_names)
```

```python
import pandas as pd
import re
import nltk
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer

nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()

def clean_and_lemmatize(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text.lower())

    words = [lemmatizer.lemmatize(word) for word in text.split() if len(word) > 2]

    return " ".join(words)

cleaned_content = [clean_and_lemmatize(doc) for doc in content]

tfidf_files = TfidfVectorizer(stop_words="english")
result_files = tfidf_files.fit_transform(cleaned_content)

feature_names_files = tfidf_files.get_feature_names_out()
df_tfidf_files = pd.DataFrame(result_files.toarray(), columns=feature_names_files)

df_tfidf_files.head(10)
```

```python
import nltk
from nltk.corpus import words
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

nltk.download('words')
english_vocab = set(w.lower() for w in words.words())

def remove_gibberish(text):
    tokens = text.lower().split()
    real_words = [w for w in tokens if w in english_vocab and len(w) > 2]
    return " ".join(real_words)

meaningful_content = [remove_gibberish(doc) for doc in content]

# 3. Re-run your TF-IDF
tfidf_files = TfidfVectorizer(stop_words="english")
result_files = tfidf_files.fit_transform(meaningful_content)

# 4. Create the clean DataFrame
feature_names_files = tfidf_files.get_feature_names_out()
df_tfidf_files = pd.DataFrame(result_files.toarray(), columns=feature_names_files)

print(f"Original columns: {len(feature_names_files)}")
df_tfidf_files.head(10)
```
