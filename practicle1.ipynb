```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = {"X":[2,2,8,5,7,6],"Y":[10,5,4,8,5,4]}
df = pd.DataFrame(data)
df.head()
```

```python
plt.scatter(df["X"],df["Y"])
plt.title("Data Points")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
clustering = KMeans(n_clusters=2)
clustering.fit(df)
df["Cluster"] = clustering.fit_predict(df)

centroids = clustering.cluster_centers_
print("Centroids:")
print(centroids)

plt.scatter(df["X"], df["Y"], c=df["Cluster"], cmap='rainbow', alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=50, marker='X', label='Centroids')

plt.title("KMeans Clustering with Midpoints")
plt.legend()
plt.show()
```

```python
np.random.seed(42)
df = pd.DataFrame({
    "X": np.random.uniform(2, 8, 100),
    "Y": np.random.uniform(4, 10, 100)
})

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

for i in range(1, 10):
    model = KMeans(n_clusters=i, n_init='auto', random_state=42)
    labels = model.fit_predict(df)
    
    centroids = model.cluster_centers_
    
    print(f"\nCentroid Coordinates for k={i}:")
    for idx, center in enumerate(centroids):
        print(f" Cluster {idx}: X={center[0]:.2f}, Y={center[1]:.2f}")

    ax = axes[i-1]
    ax.scatter(df["X"], df["Y"], c=labels, cmap='viridis', s=10, alpha=0.4)
    ax.scatter(centroids[:, 0], centroids[:, 1], c='black', s=100, marker='X')
    ax.set_title(f"k = {i}")

plt.tight_layout()
plt.show()
```

```python
data = {
    "CustomerID":[x for x in range(1, 11)],
    "AnnualIncome":[15,16,17,18,19,20,70,80,85,90],
    "SpendingScore":[39,81,6,77,40,76,94,3,60,15]
}
df = pd.DataFrame(data)

X = df[["AnnualIncome", "SpendingScore"]]

clustering = KMeans(n_clusters=3)
df["Cluster"] = clustering.fit_predict(X)

centroids = clustering.cluster_centers_

print("Centroids:")
print(centroids)

plt.scatter(df["AnnualIncome"], df["SpendingScore"], c=df["Cluster"], cmap='rainbow', alpha=0.5, s=100)
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=100, marker='X', label='Centroids')

plt.title("KMeans Clustering with Midpoints")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```
