```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
from numpy import vectorize
documents = [
    "I love eating hot pizza and cheesy burgers for dinner",
    "Pasta and pizza are my absolute favorite Italian food",
    "The new smartphone has an amazing camera and vibrant screen",
    "I desperately need to upgrade y laptop computer and phone",
    "Cooking a delicious meal with fresh ingredients like pasta"
]
```

```python
vectorizer = CountVectorizer(stop_words='english')
word_counts = vectorizer.fit_transform(documents)
feature_names = vectorizer.get_feature_names_out()

lda = LatentDirichletAllocation(n_components=3, random_state=42)
lda.fit(word_counts)

print("Discovered Topics:- ")
for topic_idx, topic in enumerate(lda.components_):
  top_words_idx = topic.argsort()[:-7:-1]
  top_words = [feature_names[i] for i in top_words_idx]
  print(f"Hidden Topic #{topic_idx + 1}: {','.join(top_words)}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)

nmf = NMF(n_components=2, random_state=42)
nmf.fit(X)

feature_names = vectorizer.get_feature_names_out()

for idx, topic in enumerate(nmf.components_):
    top_words = [feature_names[i]
                 for i in topic.argsort()[:-6:-1]]
    print(f"Topic {idx+1}: {', '.join(top_words)}")
```

```python
from sklearn.datasets import fetch_20newsgroups

newgroups = fetch_20newsgroups(subset='train', remove=('headers', 'footers', 'quotes'))
documents = newgroups.data

print(f"Loaded {len(documents)} news group articles succesfully!")
```

```python
print("PRINTING THE FIRST 3 DOCUMENTS:-")
for i, doc in enumerate(documents[:3]):
  print(f"\n[Docuent #{i+1}]")
  print(doc[:200]+"---")
```

```python
vectorizer = CountVectorizer(stop_words='english')
word_counts = vectorizer.fit_transform(documents)
feature_names = vectorizer.get_feature_names_out()

for i in range(1,5):
  lda = LatentDirichletAllocation(n_components=i, random_state=42)
  lda.fit(word_counts)

  print("Discovered Topics:- ")
  for topic_idx, topic in enumerate(lda.components_):
    top_words_idx = topic.argsort()[:-40:-1]
    top_words = [feature_names[i] for i in top_words_idx]
    print(f"Hidden Topic #{topic_idx + 1}: {','.join(top_words)}")

  print("\n\n\n")
```
